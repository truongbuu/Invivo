```python
import model
import numpy as np
from utils.data_utils import load_csv_dataset
import matplotlib.pyplot as plt
from utils.data_utils import load_csv_dataset, reshape_y_pred
import deepchem as dc
%matplotlib inline
```

## Load the data and split it 

```python
%%capture
datasets = load_csv_dataset('data.csv')
```

## Train the Model B1 (without labels information)

```python
%%capture
modelB1 = model.Model(datasets, prior = False)
mB1_train,mB1_vali = modelB1.train(epochs = 35) 
```

### Some targets have lower AUC score than other. Let's train two more models and build an ensemble to see if this is due to parameter randomness or due to the data.

```python
%%capture
modelB2 = model.Model(datasets, prior = False)
mB2_train,mB2_vali = modelB2.train(epochs = 35) 

modelB3 = model.Model(datasets, prior = False)
mB3_train,mB3_vali = modelB3.train(epochs = 35) 
```

```python
test_m1 = np.array(modelB1.tg.predict_on_generator(modelB1.data_generator(modelB1.test_data, prior_label = False)))
test_m2 = np.array(modelB2.tg.predict_on_generator(modelB2.data_generator(modelB2.test_data, prior_label = False)))
test_m3 = np.array(modelB3.tg.predict_on_generator(modelB3.data_generator(modelB3.test_data, prior_label = False)))
test_mean = (test_m1 + test_m2 + test_m3)/3
```

```python
test_predictions = reshape_y_pred(datasets[2].y, test_mean)
scoresB = dc.metrics.Metric(
            dc.metrics.roc_auc_score, np.mean, verbose = False, mode="classification").compute_metric(datasets[2].y, test_predictions, datasets[2].w, per_task_metrics= True)
```

```python
scoreB1 = modelB1.compute_score(data='test')
scoreB2 = modelB2.compute_score(data='test')
scoreB3 = modelB3.compute_score(data='test')
```

```python
print ('Average ROC-AUC model B1: ', scoreB1[0])
print ('Average ROC-AUC model B2: ', scoreB2[0])
print ('Average ROC-AUC model B3: ', scoreB3[0])
print ('Average ROC-AUC of the ensembles: ', scoresB[0])
```

### Great, so the ensemble model improves the performance a bit

```python
plt.figure(figsize=(21,10))
plt.rcParams.update({'font.size': 15})

bar_width = 0.15

plt.bar(np.arange(1,13,1) - bar_width,scoreB1[1], bar_width, color='blue',edgecolor='black', label='B1')
plt.bar(np.arange(1,13,1) , scoreB2[1], bar_width, color='green',edgecolor='black', label='B2')
plt.bar(np.arange(1,13,1) + 1*bar_width, scoreB3[1], bar_width, color='orange',edgecolor='black', label='B3')

plt.bar(np.arange(1,13,1) +  2*bar_width,scoresB[1],bar_width, color='pink',edgecolor='black', label='Ensembles')

plt.xticks(np.arange(1, 13, 1))
plt.xlabel('Target')
plt.ylabel('ROC-AUC score')
plt.ylim(ymin=0.65)
plt.legend()
plt.show()
```

### Average Test AUC-ROC score

```python
test_score_total = []
plt.figure(figsize=(10,10))
for j in range(30):
    test_score = []
    for i in range(12):
        test_score.append(modelA.compute_score(num_prior= i, data='test')[0])
    test_score_total.append(test_score)
test_score_total = np.array(test_score_total)
plt.figure(figsize=(10,5))
plt.plot(test_score_total.mean(axis = 0))
m = test_score_total.mean(axis = 0)
std = test_score_total.std(axis = 0)
plt.fill_between(np.arange(0, 12, 1.0), (m + 1.0*std).flatten()\
                 , (m - 1.0*std).flatten(),alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
plt.grid(True)
plt.ylabel('Average AUC-ROC score')
plt.xlabel('# of labels included')
plt.xticks(np.arange(0, 12, 1.0))
plt.show()
```

### Visualize Task Correlation Matrix

```python
import pandas as pd
import seaborn as sns
plt.figure(figsize=(10,5))
full_data = pd.read_csv('data.csv')
list_tasks = ['target1', 'target2','target3','target4','target5','target6',
            'target7','target8','target9','target10','target11','target12']
df = full_data[list_tasks]

corr_matrix = df.apply(lambda x: pd.factorize(x)[0]).corr(method = 'kendall', min_periods = 1).values
ax = sns.heatmap(corr_matrix, linewidth = 0.5, xticklabels= np.arange(1,13), yticklabels= np.arange(1,13))

plt.show()
```
